```python
import requests
import pandas as pd
import sqlite3
import json
```

```python
# create a variable name for the API endpoints
product_url = "https://fakestoreapi.com/products"
cart_url = "https://fakestoreapi.com/carts"
users_url = "https://fakestoreapi.com/users"
```

```python
# create a function to fetch the data and view in JSON file

def fetch_data(name, url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print("Error fecthing data from {url}")
```

```python
# create a variable to store the product JSON file and call the function

product_data = fetch_data(name = "products", url = product_url)
print(product_data[0].keys())
```

```python
# # create a variable to store the cart JSON file and call the function

cart_data = fetch_data(name = "carts", url = cart_url)
print(cart_data[0].keys())
```

```python
# create a variable to store the user JSON file and call the function

user_data = fetch_data(name = "users", url = users_url)
print(user_data[0].keys())
```

```python
# connect to sqlite database
connection = sqlite3.connect("fake_store.db")
cursor = connection.cursor()
```

```python
# create and inserting products table

def create_product_table():
    cursor.execute("""
                    CREATE TABLE IF NOT EXISTS products(
                        id INT,
                        title VARCHAR(255),
                        price float,
                        description TEXT,
                        category VARCHAR(255),
                        image TEXT,
                        rating_rate float,
                        rating_count INT          
                    );
                    """)
    connection.commit()

    
def insert_product(product_data):
    for data in product_data:
        cursor.execute(f"""INSERT INTO products (id, title, price, description, category, image, rating_rate, rating_count)
                        VALUES (?,?,?,?,?,?,?,?);
                        """, (data['id'], data['title'], data['price'], data['description'], data['category'], data['image'],
                             data['rating']['rate'], data['rating']['count']))
        connection.commit()
    
    
    
    
    
```

```python
#call function to create table
create_product_table()
```

```python
insert_product(product_data)
```
